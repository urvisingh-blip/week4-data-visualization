```python
# Week 4: Complete Data Analysis Project
# E-commerce Sales Analysis with Visualizations
# The Developers Arena Internship

import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# ============================================
# STEP 1: LOAD DATA
# ============================================
uploaded = files.upload()  # Upload sales_data.csv
df = pd.read_csv('sales_data (1).csv')
print("✅ Data loaded!")
print(f"Shape: {df.shape}")
print(df.head())

# ============================================
# STEP 2: CLEAN DATA
# ============================================
df = df.dropna()
print(f"\n✅ Data cleaned! Rows: {len(df)}")

# ============================================
# STEP 3: ANALYZE
# ============================================
total_revenue = df['Total_Sales'].sum()
avg_sale = df['Total_Sales'].mean()
best_product = df.groupby('Product')['Total_Sales'].sum().idxmax()
product_sales = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)
region_sales = df.groupby('Region')['Total_Sales'].sum().sort_values(ascending=False)
quantity_by_product = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)

print(f"\n💰 Total Revenue: ₹{total_revenue:,.2f}")
print(f"📊 Average Sale: ₹{avg_sale:,.2f}")
print(f"🏆 Best Product: {best_product}")

# ============================================
# STEP 4: CHART 1 - BAR CHART (Sales by Product)
# ============================================
plt.figure(figsize=(10, 6))
colors = ['#2196F3', '#4CAF50', '#FF9800', '#E91E63', '#9C27B0']
bars = plt.bar(product_sales.index, product_sales.values, color=colors, edgecolor='black')
plt.title('Total Sales by Product', fontsize=16, fontweight='bold')
plt.xlabel('Product', fontsize=13)
plt.ylabel('Total Sales (₹)', fontsize=13)
plt.xticks(fontsize=11)
for bar, val in zip(bars, product_sales.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 20000,
             f'₹{val:,.0f}', ha='center', fontsize=9, fontweight='bold')
plt.tight_layout()
plt.savefig('chart1_bar.png', dpi=150)
plt.show()
print("✅ Bar chart saved!")

# ============================================
# STEP 5: CHART 2 - PIE CHART (Sales by Region)
# ============================================
plt.figure(figsize=(8, 8))
colors_pie = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
plt.pie(region_sales.values, labels=region_sales.index,
        autopct='%1.1f%%', colors=colors_pie,
        startangle=140, textprops={'fontsize': 12})
plt.title('Sales Distribution by Region', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('chart2_pie.png', dpi=150)
plt.show()
print("✅ Pie chart saved!")

# ============================================
# STEP 6: CHART 3 - HORIZONTAL BAR (Quantity Sold)
# ============================================
plt.figure(figsize=(10, 6))
plt.barh(quantity_by_product.index, quantity_by_product.values,
         color=['#FF5722','#607D8B','#009688','#3F51B5','#FFC107'],
         edgecolor='black')
plt.title('Total Quantity Sold by Product', fontsize=16, fontweight='bold')
plt.xlabel('Quantity Sold', fontsize=13)
plt.ylabel('Product', fontsize=13)
for i, val in enumerate(quantity_by_product.values):
    plt.text(val + 1, i, str(val), va='center', fontsize=11, fontweight='bold')
plt.tight_layout()
plt.savefig('chart3_horizontal.png', dpi=150)
plt.show()
print("✅ Horizontal bar chart saved!")

# ============================================
# STEP 7: FINAL REPORT
# ============================================
print("\n" + "="*55)
print("📊 COMPLETE SALES ANALYSIS REPORT - WEEK 4")
print("="*55)
print(f"Total Revenue:         ₹{total_revenue:,.2f}")
print(f"Average Sale Value:    ₹{avg_sale:,.2f}")
print(f"Best Selling Product:  {best_product}")
print(f"Top Region:            {region_sales.index[0]}")
print(f"Total Transactions:    {len(df)}")
print("\n📈 INSIGHTS:")
print(f"• Laptop dominates sales contributing the most revenue")
print(f"• North region leads with highest total sales")
print(f"• Headphones and Monitor have lower sales volumes")
print(f"• Top 2 products (Laptop + Tablet) = majority of revenue")
print("="*55)
print("✅ Analysis Complete!")

# Download charts
files.download('chart1_bar.png')
files.download('chart2_pie.png')
files.download('chart3_horizontal.png')
```
